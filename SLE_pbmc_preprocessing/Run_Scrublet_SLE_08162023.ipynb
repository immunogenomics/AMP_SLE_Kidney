```python
import scanpy as sc
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import cna
import anndata as ad
import scipy.io as scio
import scipy as sci
import scrublet as scr
import scanpy as sc
import random as random
random.seed(0)
```

```python
scrub_samples = pd.read_csv('/data/srlab/ssg34/SLE_pbmc_analysis/data/scrublet/input/scrublet_samples.csv', header = None)[0].tolist()
```

```python
def scrub_sample(sample):
    matrix_path = "/data/srlab/ssg34/SLE_pbmc_analysis/data/scrublet/input/" + sample + ".mtx"
    scrub_matrix = scio.mmread(matrix_path).T.tocsc()
    cell_names = pd.read_csv('/data/srlab/ssg34/SLE_pbmc_analysis/data/scrublet/input/' + sample + '_cells.csv', header = None)
    gene_names = pd.read_csv('/data/srlab/ssg34/SLE_pbmc_analysis/data/scrublet/input/' + sample + '_genes.csv', header = None) 
    if len(cell_names) > 30:
        scrub = scr.Scrublet(scrub_matrix)
        doublet_scores, predicted_doublets = scrub.scrub_doublets()
        out = pd.DataFrame([doublet_scores, 
                             predicted_doublets,
                             [sample] * len(cell_names),
                             cell_names[0]]).transpose()
        out.columns = ["doublet_score", 'doublet_prediction', 'Scrublet_Sample', 'Cell']
        out.to_csv('/data/srlab/ssg34/SLE_pbmc_analysis/data/scrublet/output/' + sample + '_scrub_out.csv')
    else:
        out = pd.DataFrame([[float('nan')] * len(cell_names), 
                            ["NA"] * len(cell_names),
                            [sample] * len(cell_names),
                            cell_names[0]]).transpose()
        out.columns = ["doublet_score", 'doublet_prediction', 'Scrublet_Sample', 'Cell']
        out.to_csv('/data/srlab/ssg34/SLE_pbmc_analysis/data/scrublet/output/' + sample + '_scrub_out.csv')
```

```python
for i in scrub_samples: 
    scrub_sample(i)
```
